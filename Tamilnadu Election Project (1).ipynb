# 5.Visualization

```python
party_count = df['Leading Party'].value_counts()

plt.bar(party_count.index,
        party_count.values,
        color='purple')

plt.xlabel('Party')
plt.ylabel('Seats Won')
plt.title('Party Wise Winners')
plt.xticks(rotation=90)
plt.figure(figsize=(12,6))

plt.show()
```

What I Analyzed in This Bar Chart

```python
party_count = df['Leading Party'].value_counts().head(5) 
E=[0.03,0.01,0.02,0.2,0.001]
plt.figure(figsize=(10,10))
plt.pie(party_count.values, 
        labels=party_count.index,explode=E,
        autopct='%1.2f%%', labeldistance=1.1)
plt.title('Party Seat Share')
plt.show()
```

What I Analyzed in This Pie Chart

```python
plt.figure(figsize=(10,5))

plt.hist(df['Margin'], bins=10)

plt.xlabel('Winning Margin')
plt.ylabel('Frequency')
plt.title('Winning Margin Distribution')

plt.show()
```

What You Analyzed

```python
top_margin = df.sort_values(by='Margin',
                            ascending=False).head(10)

plt.figure(figsize=(10,5))

plt.bar(top_margin['Constituency'],
        top_margin['Margin'],
        color='purple')

plt.xticks(rotation=90)
plt.xlabel('Constituency')
plt.ylabel('Winning Margin')

plt.title('Top 10 Winning Margins')

plt.show()
```
